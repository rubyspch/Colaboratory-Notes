**Matplotlib**
A Python library useful for plotting to visualize data. 
Pyplot is one module of Matplotlib- provides interface of MATLAB but uses Python. Pyplot provides simple functions to add plot elements like lines, images, text, etc. to the current axes in the current figure.

Parts of a Matplotlib figure:
- Figure- A canvas that can contain one or more plots.
-Axes- a plot. x and y. Contains 2 or 3 Axis objects.
Artist- Everything you can see on the figure. E.g Text objects, Line2D objects, collection objects.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[0,5,10,15]) #pass arrays through plot method to create plot
plt.show() #invokes the plot
```

First array parameter appears as x and second appears as y

```python
plt.plot([1,2,3,4],[0,5,10,15])
plt.title('First Plot') #Makes title
plt.xlabel('X Label') #labels x
plt.ylabel('Y Label') #labels y
plt.show()
```

You can pecify the size of the figure using method figure(). Pass values as a tuple to the argument figsize:

```python
plt.figure(figsize=(5,3)) #first value specifies x, second is y.
plt.plot([1,2,3,4],[0,5,10,15]) 
plt.show()
```

Can also change colour and style of line. E.g:

```python
plt.plot([1,2,3,4],[0,5,10,15], 'go') 
plt.show()
```

can plot multiple sets on one plot:

```python
x=np.arange(1,5)
y=x**2
plt.plot([1,2,3,4],[0,5,10,15], 'go', x, y, 'r^') 
plt.show()
```

Multiple plots in one figure:
subplot() method:
- takes 3 arguments (nrows, ncols, index)

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r^')
plt.title('x y subplot')

plt.subplot(1,3,2)
plt.plot([1,2,3,4],[0,5,10,15], 'go') 
plt.title('array subplot')

plt.subplot(1,3,3)
plt.plot([2.5,3,3.5,4,4.5,5], [1,2,3,4,5,6], 'b+')
plt.title('decimal subplot')

plt.suptitle('These are my subplots')
plt.show()
```

^This method is tedious if you are making lots. So:
Use subplots() 
- takes 2 args (nrows,ncols)
- creates figure and axes to be stored in variables.
- set them by indexing rows and columns within to axes [0,1][1,0]

```python
figure_var, axes_var = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
axes_var[0,1].plot(x, x**2, 'b+')
axes_var[0,1].set_title('crosses')
axes_var[1,0].plot([1,2,3,4],[4,3,2,1], 'bo')
axes_var[1,0].set_title('dots')
```

# Bar Graphs
Method bar() which take arguments: categorical variables, their values and color (optional)


```python
groups=["A", "B", "C", "D"] 
scores=[10, 8, 6, 9]

plt.bar(groups,scores, color="orange")

#plt title, x and y labels are done the same here.
```

```python
plt.barh(groups,scores, color="orange") #barh makes horizontal
```

Stacking graphs:

have to mention index and width to stack graphs together.
legend() creates a key

```python
groupn=["A", "B", "C", "D"] 
scores1=[9, 8, 6, 9]
scores2=[10,10,9,10]
index=np.arange(4)
width=0.4

plt.bar(index, scores1, width, color="orange", label="before")
plt.bar(index+width, scores2, width, color="yellow", label="after")
plt.title('Scores before and after')

plt.ylabel("scores")
plt.xlabel("group name")
plt.xticks(index+width/2,groupn)

plt.legend(loc='best')
plt.show()
```

To vertically stack the bar graphs together, we can use an argument bottom and mention the bar graph which we want to stack below as its value:

```python
plt.bar(index, scores1, width, color="orange", label="before")
plt.bar(index, scores2, width, color="yellow", label="after", bottom=scores1)
plt.title('Scores before and after')

plt.ylabel("scores")
plt.xlabel("group name")
plt.xticks(index,groupn)

plt.legend(loc='best')
plt.show()
```

Summary list of methods:

*   plot(x-axis values, y-axis values) -
plots a simple line graph with x-axis values against y-axis values
*   show()  displays the graph
title(string)  set the title of the plot as specified by the string
*   xlabel(string)  set the label for x-axis as specified by the string
*   ylabel(string)  set the label for y-axis as specified by the string
*   figure()  used to control a figure level attributes
*   subplot(nrows, ncols, index)  Add a subplot to the current figure
*   suptitle(string)  It adds a common title to the figure specified by the string
*   subplots(nrows, ncols, figsize)  a convenient way to create subplots, in a single call. It returns a tuple of a figure and number of axes.
*   set_title(string)  an axes level method used to set the title of subplots in a figure
*   bar(categorical variables, values, color)  used to create vertical bar graphs
*   barh(categorical variables, values, color)  used to create horizontal bar graphs
*   legend(loc)  used to make legend of the graph
*   xticks(index, categorical variables)  Get or set the current tick locations and labels of the x-axis
*   pie(value, categorical variables)  used to create a pie chart
*   hist(values, number of bins)  used to create a histogram
*   xlim(start value, end value)  used to set the limit of values of the x-axis
*   ylim(start value, end value)  used to set the limit of values of the y-axis
*   scatter(x-axis values, y-axis values)  plots a scatter plot with x-axis values against y-axis values
*   axes()  adds an axes to the current figure
*   set_xlabel(string)  axes level method used to set the x-label of the plot specified as a string
*   set_ylabel(string)  axes level method used to set the y-label of the plot specified as a string
*   scatter3D(x-axis values, y-axis values)  plots a three-dimensional scatter plot with x-axis values against y-axis values
*   plot3D(x-axis values, y-axis values)  plots a three-dimensional line graph with x-axis values against y-axis values
