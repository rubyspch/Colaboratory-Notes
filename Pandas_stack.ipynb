Stacking/unstacking to change how table stacks columns. Swap columns to rows and rows to columns.

```python
import pandas as pd
```

```python
df= pd.read_excel("/content/stocks.xlsx", header=[0,1], index_col=[0]) #the data has 2 levels of header, the price/p/e and the company, so supply these as 0,1 when you read the data
df
```

to stack:

```python
df.stack() #this takes name of company from column level and transforms into  rows
```

```python
df_stacked=df.stack(level=0, dropna=True) #level arg takes the first row instead
df_stacked
```

```python
df_stacked.unstack()
```
