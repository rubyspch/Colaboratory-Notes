```python
import pandas as pd
```

This Lesson: pd.melt

```python
df=pd.DataFrame({
   "day":["Monday","Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], 
   "chicago": [20, 25, 30 , 40 , 25, 26, 27],
   "chennai": [75, 70, 76, 69, 80, 78, 69],
   "berlin":[40, 30, 39, 35, 38, 40, 33]
})
df
```

melt takes args: your df, id_vars (meaning, the column you want to keep in tact/not change)

```python
df1=pd.melt(df, id_vars="day")
df1 #you get the df transformed to format where city is inside rows.
```

```python
#can now filter down info just the daily temps for individual cities
df1[df1["variable"]=="chicago"]
```

can use another argument on df1 to change the column names, which come out as generic variable and value:
change these using var_name and value_name

```python
df2=pd.melt(df, id_vars=["day"], var_name="city", value_name="tempurature")
df2
```

```python
#filter again for temps greater than 60
df2[df2["tempurature"]>60]
```

```python
df2[df2["day"]== "Saturday"]
```
