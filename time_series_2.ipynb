Date_range function

First i will remove the dates from the csv to show how to add them using date_time function...

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/aapl.csv", parse_dates=["Date"], index_col="Date")
df.head()
```

```python
df_june=df["2017-06-30":"2017-06-01"]
df_june
```

```python
df_june=df_june.reset_index()
df_june=df_june.drop(columns="Date")
df_june #now df_june exists with the stock data but without the date column
```

#Using date_range:

```python
df_range=pd.date_range(start="6/1/2017", end="6/30/2017", freq="B") #this creates the range bw those dates with business day frequency
df_range
```

```python
df_june.set_index(df_range, inplace=True) #now we set the range as the index for df_june 
df_june
```

Benefits of datetime: your plots will include dates, easier to read. 
also can specify date to select data you want to use. 

If you want to add weekends to the data, using the info for fridays:

use asfreq to fill in holes in the date range

```python
df_june.asfreq("D", method="pad") #this pads the missing dates with the values from day before, so it carries the friday values to the weekend
```

what if you dont know the end date for the values:

```python
date_range2= pd.date_range(start="1/2/2017", periods=72, freq="B")
date_range2 # doing this gives 72 dates, excluding weekends, from 1/2/2017
```

^ you can use this to create test data to work with, by generating random numbers and putting them together with the generated dates...

```python
import numpy as np
numbers=np.random.randint(1,20, len(date_range2))
numbers # so now you have an array of random ints 1-20, the same length as the dates
```

```python
ts=pd.Series(numbers, index=date_range2) #create series from the random numbers and the generated dates
ts.head()
```
